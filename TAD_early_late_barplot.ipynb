```python
persentage = []
i = 1
for index, row in tqdm(df_tad_ut_2024.iterrows(), total=len(df_tad_ut_2024)):
    count = df_timming[
        (df_timming[0] == row[0]) & 
        (df_timming[1] >= row[1]) &  
        (df_timming[2] <= row[2]) ]
    positive_percentage, negative_percentage = (count[4] > 0).mean() * 100, (count[4] < 0).mean() * 100
    persentage.append([row[0], row[1], row[2], positive_percentage, negative_percentage])
per = 80
df_per = pd.DataFrame(persentage, columns=['chr', 'start', 'end', 'early', 'late'])
print(len(df_per[df_per['early'] >= per]))
df_per_early = df_per[df_per['early'] >= per]
print(len(df_per[df_per['late'] >= per]))
df_per_late = df_per[df_per['late'] >= per]
print(len(df_per) - len(df_per[df_per['early'] >= per]) - len(df_per[df_per['late'] >= per]))
df_per_mid = df_per[(df_per['early'] < per) & (df_per['late'] < per)]
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create the DataFrame from the array
data = [
    ['type', 'UT_2023', 'HU_2023', 'UT_2024', 'HU_2024'],
    ['early', 4304, 4346, 4132, 4374],
    ['late', 2631, 2760, 2605, 2876],
    ['mid', 462, 494, 454, 501],
    ['total', 7379, 7600, 7191, 7724]
]


# data = [['type', 'UT_2023', 'HU_2023', 'UT_2024', 'HU_2024' ],
#             ['early', 3835, 3903, 3669, 3889],
#             ['late', 2291, 2430, 2301, 2541],
#             ['mid', 1229, 1243, 1197, 1275],
#             ['total', 7335 ,7576, 7167, 7700]]


columns = data[0]  # Extract column names
rows = data[1:]    # Extract rows
df = pd.DataFrame(rows, columns=columns)

# Convert numeric columns to integers
for col in df.columns[1:]:
    df[col] = pd.to_numeric(df[col])

# Calculate percentage for each type
df_percentage = df.copy()
for col in df.columns[1:]:
    df_percentage[col] = (df[col] / df.loc[df['type'] == 'total', col].values[0]) * 100

# Remove the 'total' row for the plot
df_percentage = df_percentage[df_percentage['type'] != 'total']

# Plot the stacked bar chart
plt.figure(figsize=(10, 6))
x_labels = df.columns[1:]  # Labels from UT_2023 to HU_2024
bottom = [0] * len(x_labels)  # Initialize the bottom of stacked bars

for _, row in df_percentage.iterrows():
    values = row[1:].values  # Percentages for this type
    plt.bar(x_labels, values, bottom=bottom, label=row['type'])

    # Annotate each bar segment with its value
    for i, value in enumerate(values):
        plt.text(
            i,                  # x-position
            bottom[i] + value / 2,  # y-position (center of the bar segment)
            f"{value:.1f}%",   # Text to display
            ha="center",       # Horizontal alignment
            va="center",       # Vertical alignment
            color="white",     # Text color
            fontsize=10        # Text font size
        )
    bottom = [i + j for i, j in zip(bottom, values)]  # Update bottom for stacking

# Customize the plot
plt.title("Percentage Distribution of Each Type (Stacked)")
plt.ylabel("Percentage")
plt.xlabel("Samples")
plt.legend(title="Type")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.ylim(0, 100)  # Ensure bars are scaled to 100%
plt.tight_layout()

# Show the plot
plt.savefig('../pics/TAD_distribution_boundaries.svg', dpi=1200, format='svg')
plt.show()
```
