```python
from matplotlib_venn import venn2
import matplotlib.pyplot as plt

# ======== EDIT YOUR VALUES HERE ========
set1 = 84627  # only in set 1
set2 = 48194   # only in set 2
common = 40727  # shared between set1 & set2
unique_set1 = set1 - common
unique_set2 = set2 - common

# ======== COLORS (editable) ========
color1 = "#ff0000"   # gray
color2 = "#0000ff"   # red

# ======== CALCULATE PERCENTAGES ========
total = unique_set1 + common
p1 = unique_set1 / total * 100
p2 = unique_set2 / total * 100
pc = common / total * 100

# ======== SET GLOBAL FONT ========
# plt.rcParams["font.family"] = "Arial"
# plt.rcParams["font.size"] = 10

# ======== PLOT ========
plt.figure(figsize=(5,5))
v = venn2(subsets=(unique_set1, unique_set2, common),
          set_labels=("Set 1", "Set 2"),
          set_colors=(color1, color2),
          alpha=0.6)

# Outline + aesthetics
for patch in v.patches:
    if patch:
        patch.set_edgecolor("black")
        patch.set_linewidth(1.2)

# ======== ADD COUNTS + PERCENTAGES ========
if v.get_label_by_id('10'):
    v.get_label_by_id('10').set_text(f"{unique_set1}\n({p1:.1f}%)")
if v.get_label_by_id('01'):
    v.get_label_by_id('01').set_text(f"{unique_set2}\n({p2:.1f}%)")
if v.get_label_by_id('11'):
    v.get_label_by_id('11').set_text(f"{common}\n({pc:.1f}%)")

# # ======== STYLE TEXT LABELS ========
# for label in v.set_labels:
#     if label:
#         label.set_fontsize(10)
#         label.set_fontfamily("Arial")
# for label in v.subset_labels:
#     if label:
#         label.set_fontsize(10)
#         label.set_fontfamily("Arial")

# plt.title("Venn Diagram — 2 Sets", fontsize=10, family="Arial")

# ======== SAVE + SHOW ========
plt.tight_layout()
plt.savefig("../fig_1/venn_diagram_HU_acolin.svg")
plt.show()

print("✅ Saved as 'venn_diagram.svg' at 1200 DPI (Arial, font size 10).")
```
